```python
from __future__ import absolute_import, division, print_function, unicode_literals
import os
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import cv2
```

```python
import tensorflow_datasets as tfds
tfds.disable_progress_bar()
```

```python
#Show image in a testset

DATA_DIR = 'data/PetImages'
Categories = ['Dog', 'Cat']

for category in Categories:
    path = os.path.join(DATA_DIR, category)
    for image in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,image))
        plt.imshow(img_array)
        plt.title("pupper")
        plt.show()
        break
    break
```

```python
IMG_SIZE = 150
new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(new_array)
plt.title("phat pupper")
plt.show()
```

```python
#Create Training Data


def create_training_data():
    training_data = []
    for category in Categories:
        count = 0
        classification = Categories.index(category)
        
        path = os.path.join(DATA_DIR, category)
        for image in os.listdir(path):
            
            try:
                img_array = cv2.imread(os.path.join(path,image))
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                training_data.append([new_array, classification])
                count += 1
                if count ==4000:
                    break
            except Exception as e:
                pass
    
    return training_data

train_data = create_training_data()
print(len(train_data))
```

```python
#Shuffle Data
random.shuffle(train_data)
print(len(train_data))
```

```python
X = []
y = []


for image, label in train_data:
    X.append(image)
    y.append(label)


X = np.asarray(X).reshape(-1, IMG_SIZE, IMG_SIZE, 3)

#Normalize X
X = X/np.mean(X)
print(X.shape)
```

```python
#Hand made VGG16 style CNN
model = Sequential()
model.add(Conv2D(32, (3,3), input_shape = X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(128, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(128))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X,y, batch_size = 32, epochs = 15, validation_split=.2)
```
